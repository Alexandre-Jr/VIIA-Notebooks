# **2. Insira imagens**

**2.1. Criar pasta "images" e "results"**

Apenas execute o código.

```python
files = [f for f in os.listdir()]
there_is_imagesFolder = False
there_is_resultsFolder = False

for file in files:
  if file == "results":
    there_is_resultsFolder = True
  if file == "images":
    there_is_imagesFolder = True

if not there_is_resultsFolder:
  os.mkdir("results")
if not there_is_imagesFolder:
  os.mkdir("images")
```

**2.2 Insira a sua imagem em "./images"**

Abra a aba do arquivos e encontre a pasta "images", entre e insirá a sua(s) imagem(s).

**Cheque se a sua imagem foi inserida no local correto, se foi, aparecerá aqui:**

```python
# Captura e lista as imagens em content/images
files = [f for f in os.listdir(path="./images")]

print(f'Na sua pasta existe: {files}')

# Exibe a imagem
for file in files:
  if(file[0] != '.'):
    image = plt.imread(f'./images/{file}')
    plt.imshow(image)
    plt.axis('off')
    plt.show()

```

# **3. Classificação de imagems**

Será exibida a imagem do resultado da classificação, e a imagem resultante pode ser baixada ou encontrada em formato de arquivo na pasta "results"

```python
# Carrega o modelo
model = YOLO(modelclass)

# Captura e lista as imagens em content/images
files = [f for f in os.listdir(path="./images")]

print(f'Na sua pasta existe: {files}')

# Exibe a imagem
for file in files:
  if(file[0] != '.'):
    result = model(f'./images/{file}')[0]
    im_bgr = result.plot();
    im_rgb = im_bgr[..., ::-1]

    plt.imshow(im_rgb)
    plt.axis('off')
    plt.show()

    # Salva o resultado
    cv2.imwrite(f'./results/class-{file}', im_bgr)
    print(f"Salvo: ./results/class-{file}")
```

# **4. Segmentação de instâncias em imagems**

**4.1. Segmentação de instâncias**

Será exibida a imagem do resultado da segmentação de instâncias, e a imagem resultante pode ser baixada ou encontrada em formato de arquivo na pasta "results"

```python
# Carrega o modelo
model = YOLO(modelseg)

# Captura e lista as imagens em content/images
files = [f for f in os.listdir(path="./images")]

print(f'Na sua pasta existe: {files}')

# Exibe a imagem
for file in files:
  if(file[0] != '.'):
    result = model(f'./images/{file}')[0]
    im_bgr = result.plot();
    im_rgb = im_bgr[..., ::-1]

    plt.imshow(im_rgb)
    plt.axis('off')
    plt.show()

    # Salva o resultado
    cv2.imwrite(f'./results/seg-{file}', im_bgr)
    print(f"Salvo: ./results/seg-{file}")
```
